```python
import numpy as np
import scipy.io as sci
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(1)
n = 10
x = np.zeros(n)
g = np.zeros(n)
t = np.zeros(n)

for i in np.arange(n):
    x[i] = np.random.uniform(0, 1)
    g[i] = np.random.normal(0, 1)
    t[i] = np.sin(2 * np.pi * x[i]) + 0.1*g[i]

```

```python
f = np.linspace(0, 1, 100)

fig = plt.figure()
plt.scatter(x, t)
plt.plot(f, np.sin(2*f*np.pi), 'r')
plt.legend(['sin(2x*pi)', 'data points'])
plt.xlabel('x')
plt.ylabel('t')
```

```python
phi = np.column_stack((np.ones(n), x, x**2))
test = np.matmul(np.linalg.inv(np.matmul(phi.T, phi)), phi.T)
wML = np.dot(test, t.T)
```
